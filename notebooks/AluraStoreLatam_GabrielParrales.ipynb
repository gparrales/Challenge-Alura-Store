### Importación de datos



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Calculamos la facturación de cada tienda en base a la variable "Precio"
facturacion = tienda['Precio'].sum()
facturacion1 = tienda2['Precio'].sum()
facturacion2 = tienda3['Precio'].sum()
facturacion3 = tienda4['Precio'].sum()

# Creamos un dataframe para agrupar los valores
datos = {
    'Tiendas' : ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ventas' : [facturacion, facturacion1, facturacion2, facturacion3]
}
df_ventas = pd.DataFrame(datos)
pd.options.display.float_format = '{:.2f}'.format
df_ventas

```

```python
plt.bar(df_ventas['Tiendas'], df_ventas['Ventas'], width=0.4)
plt.title('Ventas por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ventas')
for i, value in enumerate(df_ventas['Ventas']):
    plt.text(i, value + 5, str(value), ha='center', fontsize=6)
plt.xticks(fontsize=10)
plt.yticks(fontsize=6)
plt.yticks([])
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.show()
```

```python
plt.pie(df_ventas['Ventas'], labels=df_ventas['Tiendas'], autopct='%1.1f%%', startangle=90)
plt.title('Porcentaje de Ventas por Tienda')
plt.axis('equal')
```

# 3. Calificación promedio de la tienda


```python
# Se obtiene la calificación promedio de cada tienda
calificacion = tienda['Calificación'].mean()
calificacion2 = tienda2['Calificación'].mean()
calificacion3 = tienda3['Calificación'].mean()
calificacion4 = tienda4['Calificación'].mean()

# Creamos un dataframe para agrupar los valores
cal = {
    'Tiendas' : ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificación Promedio' : [calificacion, calificacion2, calificacion3, calificacion4]
}
df_cal = pd.DataFrame(cal)
df_cal.sort_values(by='Calificación Promedio', ascending=False)
```

```python
plt.bar(df_cal['Tiendas'], df_cal['Calificación Promedio'], width=0.4)
plt.title('Calificación Promedio por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Calificación Promedio')
for i, value in enumerate(df_cal['Calificación Promedio']):
    plt.text(i, value + 0.1, str(round(value, 2)), ha='center', fontsize=6)
plt.xticks(fontsize=10)
plt.yticks(fontsize=6)
plt.yticks([])
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
```

# 5. Envío promedio por tienda

```python
# Se calcula el costo de envío promedio para cada tienda
costo_envio = tienda['Costo de envío'].mean()
costo_envio2 = tienda2['Costo de envío'].mean()
costo_envio3 = tienda3['Costo de envío'].mean()
costo_envio4 = tienda4['Costo de envío'].mean()

# Se crea un dataframe con los valores obtenidos

ce = {
    'Tiendas' : ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Costo de Envío Promedio' : [costo_envio, costo_envio2, costo_envio3, costo_envio4]
}
df_costos = pd.DataFrame(ce)
df_costos = df_costos.sort_values(by='Costo de Envío Promedio', ascending=False)
df_costos



```

```python
plt.bar(df_costos['Tiendas'], df_costos['Costo de Envío Promedio'], width=0.4)
plt.title('Costo de envío promedio por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Costo de envío')
for i, value in enumerate(df_costos['Costo de Envío Promedio']):
    plt.text(i, value + 5, str(round(value,2)), ha='center', fontsize=8)
plt.xticks(fontsize=10)
plt.yticks(fontsize=6)
plt.yticks([])
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.show()
```
